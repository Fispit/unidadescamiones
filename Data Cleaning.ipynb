```python
#Applying functions to only get the geoshape points as an array
test['geo_shape']=test['geo_shape'].map(lambda x: x.lstrip('{"type": "Polygon", "coordinates": ').rstrip('}'))
test['geo_shape'] = test['geo_shape'].map(lambda x: x[1:-1])
```

```python
alcaldias=test[["id","nomgeo","geo_shape"]]
alcaldias=alcaldias.rename(columns={"nomgeo":"alcaldia"})
alcaldias
```

```python
alcaldias.to_csv("alcaldias.csv",index=False)
```

```python
test2=data_df
test2["alcaldia"]=""
```

```python
#funcion para incluir la alcaldia en los datos de camiones
def loc_alcaldias(x):
    lat=x.position_latitude
    lon=x.position_longitude
    
    for x in range(0,len(alcaldias.id)):
        polygon=ast.literal_eval(alcaldias.geo_shape[x])
        poly_path = mplPath.Path(np.array(polygon))
        point = (lon, lat)
        inalc=poly_path.contains_point(point)
        if inalc:
            return alcaldias.alcaldia[x]
```

```python
test2["alcaldia"]=test2.apply(loc_alcaldias,axis=1)
```
